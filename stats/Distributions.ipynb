```python
normal = pd.DataFrame({'rand' : np.random.normal(25, 20, 100)})
plt.figure(figsize= (20, 10))
plt.hist(normal['rand'])
plt.axvline(normal['rand'].mean(), color='y', linestyle='solid', linewidth=2)
plt.axvline(normal['rand'].mean() + normal['rand'].std(), color ='y', linestyle='dashed', linewidth=2)
plt.axvline(normal['rand'].mean() - normal['rand'].std(), color ='y', linestyle='dashed', linewidth=2)
plt.show()
normal.describe()
```

```python
#The descriptive stats tell us quite a lot about this distribution. 
#By knowing the mean we know where most of the values are clumped and 
#the standard deviation gives us a good picture of where all the data is
```

```python
gumbel = pd.DataFrame({'rand' : np.random.gumbel(45, 8, 100)})
plt.figure(figsize= (20, 10))
plt.hist(gumbel['rand'])
plt.axvline(gumbel['rand'].mean(), color='y', linestyle='solid', linewidth=2)
plt.axvline(gumbel['rand'].mean() + gumbel['rand'].std(), color ='y', linestyle='dashed', linewidth=2)
plt.axvline(gumbel['rand'].mean() - gumbel['rand'].std(), color ='y', linestyle='dashed', linewidth=2)
plt.show()
gumbel.describe()
```

```python
# Here the mean still gives us a good bit of information about the distribution
# The standard deviation gives us a less clear picture of the data due to the 'fat tail'
```

```python
binomial = pd.DataFrame({'rand' : np.random.binomial(25, .7, 100)})
plt.figure(figsize= (20, 10))
plt.hist(binomial['rand'])
plt.axvline(binomial['rand'].mean(), color='y', linestyle='solid', linewidth=2)
plt.axvline(binomial['rand'].mean() + binomial['rand'].std(), color ='y', linestyle='dashed', linewidth=2)
plt.axvline(binomial['rand'].mean() - binomial['rand'].std(), color ='y', linestyle='dashed', linewidth=2)
plt.show()
binomial.describe()
```

```python
# The mean matches the most likely outcome very closely here
# The data is not very neatly described by the standard deviation 
# moreover as p changes the distribution would change a lot
```

```python
gamma = pd.DataFrame({'rand' : np.random.gamma(20, .5, 100)})
plt.figure(figsize= (20, 10))
plt.hist(gamma['rand'])
plt.axvline(gamma['rand'].mean(), color='y', linestyle='solid', linewidth=2)
plt.axvline(gamma['rand'].mean() + gamma['rand'].std(), color ='y', linestyle='dashed', linewidth=2)
plt.axvline(gamma['rand'].mean() - gamma['rand'].std(), color ='y', linestyle='dashed', linewidth=2)
plt.show()
gamma.describe()
```

```python
# The mean shows us where most of the data are clustered in this distribution.
# Which in this case would mean when the event is most likely to happen
# This is another distribution with a 'fat tail' 
# standard deviations are less helpful because this distribution is less dense than a normal distribution.
```
