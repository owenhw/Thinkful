```python
For two of the following distributions:

Normal
Binomial
Gamma
Poisson
Create a line chart plotting the sample mean as more and more trials are conducted, up to 1,000 trials.

Randomly sample the mean 1,000 times and plot the resulting frequency of sample means as a histogram.

When you are finished compare your results to this.
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def large_numbers(title, dist, mean, sigma, n):
    ## declare variables
    output = []
    
    ## determine which distribution we are sampling and run the samples
    if dist == 'normal':
        for trials in range(1, n):
            trial = np.random.normal(mean, sigma, trials)
            trials_mean = trial.mean()
            output.append(trials_mean)
            
    elif dist == 'binomial':
        for trials in range(1, n):
            trial = np.random.binomial(mean, sigma, trials)
            trials_mean = trial.mean()
            output.append(trials_mean)
            
    elif dist == 'poisson':
        for trials in range(1, n):
            trial = np.random.poisson(sigma, trials)
            trials_mean = trial.mean()
            output.append(trials_mean)
            
    elif dist == 'gamma':
        for trials in range(1, n):
            trial = np.random.gamma(mean, sigma, trials)
            trials_mean = trial.mean()
            output.append(trials_mean)
            
    else:
        return 'Bad input'
    
    ## return a plot of the data
    df = pd.DataFrame({'Trials' : output})   
    
    
    plt.figure(figsize = (20, 10))
    plt.plot(df['Trials'])
    plt.title(title, size='xx-large')
    plt.xlabel('Trials', size='xx-large')
    plt.ylabel('Mean of Sample', size='xx-large',)
    plt.legend(['Mean'], fontsize='xx-large')
    return plt.show()
```

```python
def central_limit(title, dist, mean, sigma, n):
    ## declare variables
    means = []

    ## determine which distribution we are sampling and run the samples
    if dist == 'normal':
        for trials in range(n):
            means.append(np.mean(np.random.normal(mean, sigma, n)))
            
    elif dist == 'binomial':
        for trials in range(n):
            means.append(np.mean(np.random.binomial(mean, sigma, n)))
                   
    elif dist == 'poisson':
        for trials in range(n):
            means.append(np.mean(np.random.poisson(sigma, n)))
                        
    elif dist == 'gamma':
        for trials in range(n):
            means.append(np.mean(np.random.gamma(mean, sigma, n)))
                       
    else:
        return 'Bad input'
    
    ## return a histogram of the data
    plt.figure(figsize = (20, 10))
    plt.hist(means)
    plt.title(title, size='xx-large')
    plt.xlabel('Mean', size='xx-large')
    plt.ylabel('Frequency', size='xx-large',)
    return plt.show()
```

```python
large_numbers('Mean of Samples of a Gamma Distribution', 'gamma', 23, 11, 1000)
```

```python
central_limit('Central limit of a Gamma Distribution', 'gamma', 23, 11, 1000)
```

```python
large_numbers('Mean of Samples of a Poisson Distribution', 'poisson', None, 3, 1000)
```

```python
central_limit('Central Limit of a Poisson Distribution', 'poisson', None, 3, 1000)
```

```python
large_numbers('Mean of Samples of a Binomial Distribution', 'binomial', 2, .8, 1000)
```

```python
central_limit('Central Limit of a Binomial Distribution', 'binomial', 2, .8, 1000)
```

```python
large_numbers('Mean of Samples of a Normal Distribution', 'normal', 20, 13, 1000)
```
