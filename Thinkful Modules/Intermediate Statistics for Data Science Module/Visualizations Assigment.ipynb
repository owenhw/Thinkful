```python
# To keep the page organized do all imports here
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
sns.set()

# suppress warnings (warnings do not prevent the code from executing)
import warnings
warnings.filterwarnings("ignore")
```

```python
houseprices = pd.read_csv('https://tf-assets-prod.s3.amazonaws.com/tf-curric/data-science/houseprices.csv')
houseprices.info()
```

```python
# create distribution of saleprice
plt.figure(figsize = (20, 10))
ax = sns.distplot(houseprices['saleprice'], bins=100)
plt.xlabel('Saleprice')
plt.ylabel('')
plt.title('Distribution of saleprice Values')
plt.show()
```

```python
# create distribution of lotarea
plt.figure(figsize= (20, 10))
bx = sns.distplot(houseprices['lotarea'], bins=70)
plt.xlabel('Lotarea')
plt.title('Distribution of lotarea Values')
```

```python
# Use a boxplot to illustrate how overallcond affects the sale price.
plt.figure(figsize = (20, 6))
sns.boxplot(y="saleprice",  x = "overallcond", data=houseprices)
plt.title ('Affects of Overall Condition on Saleprice')
plt.xlabel('Overall Condition')
plt.ylabel('Saleprice')
plt.show()
```

```python
# On the same boxplot, highlight the difference between having central air conditioning or not.
# Use a boxplot to illustrate how overallcond affects the sale price.
plt.figure(figsize = (20, 6))
sns.boxplot(y="saleprice", hue='centralair', x = "overallcond", data=houseprices)
plt.title ('Affects of Overall Condition on Saleprice')
plt.xlabel('Overall Condition')
plt.ylabel('Saleprice')
plt.show()
```

```python
# Is there a correlation between lotarea and the saleprice?
                            
plt.figure(figsize=(20, 5))
sns.boxplot(x='lotarea', y='saleprice', data=houseprices)
plt.title('Distributions of Saleprice for a Given Lot Area')
plt.xlabel('Lot Area')
plt.ylabel('Saleprice')
plt.show()
```

```python
houseprices['lotarea'].corr(houseprices['saleprice'])
```
