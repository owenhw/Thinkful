```python
postgres_user = 'dsbc_student'
postgres_pw = '7*.8G9QH21'
postgres_host = '142.93.121.174'
postgres_port = '5432'
postgres_db = 'weatherinszeged'

engine = create_engine('postgresql://{}:{}@{}:{}/{}'.format(
    postgres_user, postgres_pw, postgres_host, postgres_port, postgres_db))

query1 = '''
SELECT
  *
FROM
  weatherinszeged
'''

weather_df = pd.read_sql_query(query1, con=engine)
engine.dispose()
```

```python
weather_df['windspeed_x_humidity'] = weather_df['windspeed'] * weather_df['humidity']
```

```python
for column in ['windspeed_x_humidity', 'temperature', 'windspeed', 'humidity']:
    weather_df.loc[:, column] = winsorize(weather_df.loc[:, column], (0.05, 0.05))
```

```python
weather_1_X = weather_df[['windspeed', 'humidity']]
weather_y = weather_df['apparenttemperature'] - weather_df['temperature']
```

```python
weather_1_X = sm.add_constant(weather_1_X)

weather_1_results = sm.OLS(weather_y, weather_1_X).fit()

weather_1_results.summary()
```

```python
weather_2_X = weather_df[['windspeed_x_humidity', 'windspeed', 'humidity']]
```

```python
weather_2_X = sm.add_constant(weather_2_X)

weather_2_results = sm.OLS(weather_y, weather_2_X).fit()

weather_2_results.summary()
```

```python
weather_3_X = weather_df[['visibility', 'windspeed', 'humidity']]

weather_3_X = sm.add_constant(weather_3_X)

weather_3_results = sm.OLS(weather_y, weather_3_X).fit()

weather_3_results.summary()
```
