```python
from dask import delayed
from dask.distributed import Client, progress
import dask.dataframe as dd
```

```python
@delayed

def sum_up(l): #takes a list of numbers as input
    return sum(l)

@delayed
def cube(n):
    return n**3

total = []
for i in range(1,10):
    total.append(cube(i))
```

```python
result = sum_up(total)
```

```python
result.visualize()
```

```python
result.compute()
```

```python
# first two fibonacci numbers
fibonacci_nums = [1,1]

@delayed
# returns the nth fibonacci number
def get_nth_fibonacci_number(n):
    if n == 1:
        return fibonacci_nums[0]
    elif n == 2:
        return fibonacci_nums[1]
    else:
        return get_nth_fibonacci_number(n-1).compute() + get_nth_fibonacci_number(n-2).compute()

answer = get_nth_fibonacci_number(10)
```

```python
answer.visualize()
```

```python
answer.compute()
```
